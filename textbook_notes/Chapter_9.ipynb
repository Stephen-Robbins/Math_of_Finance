# Volatility Smile

## Implied Volatility

The implied volatility $\sigma_{imp}$ is calculated by inverting the Black-Scholes formula:

$$
\sigma_{imp} = BS^{-1}(S_0, K, r, T, C)
$$

where $S_0$ is the current asset price, $K$ is the strike price, $r$ is the risk-free rate, $T$ is the time to expiration, and $C$ is the market price of the option.

## Volatility Smile Pattern

The volatility smile refers to the observed pattern of implied volatilities $\sigma_{imp}$ across different strike prices $K$ for options with the same underlying asset and expiration date $T$.

$$
\sigma_{imp} = f(K)
$$

This pattern forms a smile-shaped curve when plotted against the strike price.

## Volatility Behavior

Empirical observations suggest that volatility exhibits certain behaviors:

### Mean Reversion

Volatility tends to revert to a long-term mean level over time. This can be modeled using a mean-reverting process:

$$
d\sigma_t = \kappa(\theta - \sigma_t)dt + \xi dW_t
$$

where $\kappa$ is the mean-reversion speed, $\theta$ is the long-term mean volatility, and $\xi$ is the volatility of volatility.

### Volatility Clustering

Periods of high (low) volatility tend to be followed by periods of high (low) volatility, exhibiting momentum or clustering behavior.

### Leverage Effect

For equity options, a decrease in the underlying stock price tends to increase its leverage and volatility, contributing to the volatility smile pattern.

```python
import numpy as np
import scipy.stats as stats
from scipy.optimize import brentq
import matplotlib.pyplot as plt
import pandas as pd

# Define the Black-Scholes price for a European call option
def black_scholes_call(S0, E, T, r, sigma):
    d1 = (np.log(S0 / E) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    return S0 * stats.norm.cdf(d1) - E * np.exp(-r * T) * stats.norm.cdf(d2)

# Function to calculate implied volatility
def implied_volatility_call(S0, E, T, r, market_price):
    def difference_in_price(sigma):
        return black_scholes_call(S0, E, T, r, sigma) - market_price
    implied_vol = brentq(difference_in_price, 0.01, 3)
    return implied_vol

# Market data
S0 = 189
r = 0.0531
T = (pd.to_datetime('2025-01-17') - pd.to_datetime('2024-02-07')).days / 365.25

# Options data
options_data = {
    'Strike Price': [165, 170, 175, 180, 185, 190, 195, 200, 205, 210],
    'Current Price': [36.56, 32.97, 29.73, 25.75, 22.70, 19.86, 17.00, 14.50, 12.61, 10.23]
}

# Calculate implied volatility for each option
implied_vols = []
for E, market_price in zip(options_data['Strike Price'], options_data['Current Price']):
    iv = implied_volatility_call(S0, E, T, r, market_price)
    implied_vols.append(iv)

# Plot the implied volatility smile
plt.figure(figsize=(10, 5))
plt.plot(options_data['Strike Price'], implied_vols, marker='o')
plt.xlabel('Strike Price')
plt.ylabel('Implied Volatility')
plt.title('Implied Volatility Smile for AAPL Options')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.stats import norm
from scipy.optimize import newton

# Black-Scholes formula for European call option
def call_option_price(S, K, T, r, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    call_price = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    return call_price

# Black-Scholes formula for European put option
def put_option_price(S, K, T, r, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    put_price = K * np.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)
    return put_price

# Objective function to find the root (implied volatility)
def implied_volatility_objective(sigma, S, K, T, r, market_price, option_type='call'):
    if option_type == 'call':
        return call_option_price(S, K, T, r, sigma) - market_price
    elif option_type == 'put':
        return put_option_price(S, K, T, r, sigma) - market_price

# Function to calculate implied volatility
def calculate_implied_volatility(S, K, T, r, market_price, option_type='call'):
    # Initial guess for volatility
    sigma_initial_guess = 0.2
    # Use Newton-Raphson method to find the root of the objective function
    implied_vol = newton(implied_volatility_objective, sigma_initial_guess, args=(S, K, T, r, market_price, option_type))
    return implied_vol

# Example usage:
S = 100  # Current stock price
K = 100  # Strike price
T = 1    # Time to expiration in years
r = 0.05 # Risk-free interest rate
market_price = 10  # Market price of the option

# Calculate implied volatility for a call option
implied_vol_call = calculate_implied_volatility(S, K, T, r, market_price, option_type='call')
print(f"Implied Volatility for Call Option: {implied_vol_call:.6f}")

# Calculate implied volatility for a put option
implied_vol_put = calculate_implied_volatility(S, K, T, r, market_price, option_type='put')
print(f"Implied Volatility for Put Option: {implied_vol_put:.6f}")
```

### Volatility Harvesting Strategies

#### Straddle/Strangle

$$
\begin{aligned}
\text{Straddle Payoff} &= \max(S_T - K, 0) + \max(K - S_T, 0) \\
\text{Strangle Payoff} &= \max(S_T - K_C, 0) + \max(K_P - S_T, 0)
\end{aligned}
$$

Straddles and strangles involve buying a call and a put option (straddle) or a higher strike call and a lower strike put (strangle). These strategies profit from high realized volatility, as the options become more valuable when the underlying asset price moves significantly in either direction.

#### Volatility Spread

$$
\begin{aligned}
\text{Long Volatility Spread} &= \nu_C(K_1, \sigma_1) - \nu_C(K_2, \sigma_2) \\
\text{Short Volatility Spread} &= \nu_P(K_1, \sigma_1) - \nu_P(K_2, \sigma_2)
\end{aligned}
$$

Volatility spreads involve taking offsetting positions in options with different strike prices and/or maturities, exploiting the difference in their vega exposures. Long volatility spreads profit from volatility increases, while short volatility spreads benefit from volatility decreases.

#### Variance Swap

$$
\text{Variance Swap Payoff} = N(\text{Realized Variance} - \text{Strike Variance})
$$

Variance swaps are over-the-counter derivatives that allow traders to speculate on the future realized variance of an underlying asset. The payoff is based on the difference between the realized variance and the strike variance, enabling traders to profit from volatility mispricing.

#### Volatility Arbitrage

$$
\begin{aligned}
\sigma_{\text{implied}} &\neq \sigma_{\text{expected}} \\
\text{Arbitrage Profit} &= \text{Option Premium} - \text{Hedging Cost}
\end{aligned}
$$

Volatility arbitrage involves identifying and exploiting discrepancies between the implied volatility from option prices and the expected future volatility. Traders can construct delta-neutral portfolios to capture the difference between the option premium and the cost of hedging, profiting from volatility mispricing.



### Problem 1

As a United States Senator in early 2020, you have confidential information about an impending pandemic that is likely to cause significant market volatility. To capitalize on the expected increase in market volatility without directly buying put options—which could be construed as insider trading—you decide to employ a strategy that involves buying and selling call options on the S&P 500 Index  to create a delta-neutral position.

#### Background Information:

- The S&P 500 Index is currently at 3,200 points.
- You are considering 6-month European call options on the S&P 500 with a strike prices rainging between 3000-3500 in increments of 100.
- The call options are trading with an implied volatility of 25%.
- The continuously compounded risk-free interest rate is 1% per annum.

You can only buy/sell 100 of each option. What is the maximum profit you can achieve while maintaining a delta neutral portfolio.


```python
from scipy.optimize import linprog
import numpy as np
from scipy.stats import norm

# Given parameters
S0 = 3200  # Current S&P 500 Index level
Ks = np.array([3000, 3100, 3200, 3300, 3400, 3500])  # Strike prices
r = 0.01  # Risk-free interest rate
T = 0.5  # Time to expiration in years
sigma = 0.25  # Implied volatility (25%)


# Define the Black-Scholes call option pricing formula
def black_scholes_call(S, K, T, r, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    call_price = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    return call_price

# Define the Delta of a call option
def delta_call(S, K, T, r, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    delta = norm.cdf(d1)
    return delta

# Calculate deltas and price differences for options with different volatilities
deltas = np.array([delta_call(S0, K, T, r, sigma) for K in Ks])

#Want to minimize the cost of our portfolio
c  = np.array([ black_scholes_call(S0, K, T, r, sigma) for K in Ks])

print(deltas, c)
# Constraints
# 1. Delta neutrality: The sum of deltas * number of options for each strike must be 0
A_eq = deltas.reshape(1, -1)
b_eq = [0]


x_bounds = [(-100, 100) for _ in Ks] 

# Solve the linear programming problem with cost constraint
result = linprog(c,  A_eq=A_eq, b_eq=b_eq, bounds=x_bounds, method='highs')


if result.success:
    # Extract the optimal number of options for each strike price
    optimal_quantities = result.x
    max_profit = -result.fun  # Negative because we minimized the negative profit
    print("Optimal number of options for each strike price:", optimal_quantities)
    print("Maximum profit:", max_profit)
else:
    print("Optimization was not successful. Error:", result.message)
```

## Stochastic Volatility Models

Stochastic volatility models are an extension of the Black-Scholes model that allow for the volatility of the asset to be a stochastic process, rather than a constant. This can help to better capture the observed volatility smile in the market.

### The Heston Model

The Heston model is a popular stochastic volatility model that assumes the asset price $S_t$ and its variance $\nu_t$ follow the dynamics:

1. **Asset Price Dynamics**: $dS_t = \mu S_t dt + \sqrt{\nu_t} S_t dW^S_t$
2. **Volatility Dynamics**: $d\nu_t = \kappa (\theta - \nu_t) dt + \xi \sqrt{\nu_t} dW^{\nu}_t$

where:
- $\mu$ is the expected return rate of the stock
- $\kappa$ is the rate of mean reversion for the variance
- $\theta$ is the long-run average price variance
- $\xi$ is the volatility of the volatility (vol of vol)
- $dW^S_t$ and $dW^{\nu}_t$ are two Wiener processes with correlation $\rho$

The Heston model allows for the volatility to be mean-reverting and have its own source of randomness, which can help to capture the volatility smile.

### The Rough Bergomi Model

The rough Bergomi model is a more recent stochastic volatility model that incorporates rough paths, which can help to better fit the observed volatility surface. The model is given by:

1. **Log-Stock Price Dynamics**: $dX_t = -\frac{1}{2} V_t dt + \sqrt{V_t} dW_t$
2. **Variance Dynamics**: $V_t = \xi_0(t) \mathcal{E} \left( \sqrt{2H} \int_0^t (t-s)^{H-\frac{1}{2}} dZ_s \right)$

where:
- $X_t$ represents the log of the stock price
- $V_t$ is the variance process
- $W_t$ and $Z_t$ are Brownian motions with correlation $\rho$
- $H$ is the Hurst parameter, which reflects the roughness of the paths
- $\xi_0(t)$ is the initial forward variance curve
- $\mathcal{E}$ is the stochastic exponential

The rough Bergomi model uses a fractional Brownian motion to drive the variance process, which can capture the rough nature of volatility and help to fit the volatility surface more accurately.

### Local Volatility Models

Local volatility models are another class of models that aim to fit the volatility smile. These models assume that the volatility is a deterministic function of the asset price and time, i.e., $\sigma(S_t, t)$. The dynamics of the asset price are given by:

$dS_t = \mu S_t dt + \sigma(S_t, t) S_t dW_t$

Local volatility models can be calibrated to fit the observed volatility surface exactly, but they do not provide a good explanation for the source of the smile and can lead to unrealistic dynamics.

# Shannon's Demon


## Setup
- Consider a game where the wealth can either double (2x) or halve (0.5x) at each time step
- The probability of doubling is $p=0.5$, and the probability of halving is also $1-p=0.5$
- The game is assumed to be memoryless (i.e., each time step is independent)

## Rebalancing Strategy
- Start with an initial wealth $W_0$ divided equally between the risky asset and cash
- At each time step, rebalance the portfolio to maintain a 50/50 allocation between the asset and cash
- The rebalanced portfolio value is given by:
  $W_t = \frac{1}{2}(W_{t-1} \cdot 2 + W_{t-1}) = \frac{3}{2}W_{t-1}$ for a doubling event, and
  $W_t = \frac{1}{2}(W_{t-1} \cdot 0.5 + W_{t-1}) = \frac{3}{4}W_{t-1}$ for a halving event

## Growth Rate
- The long-term growth rate of the rebalanced portfolio is given by:
  $g = \frac{1}{2}\log(1.5) + \frac{1}{2}\log(0.75) \approx 0.0566$
- This growth rate is always greater than the average growth rate of the asset itself, which is:
  $g_{\text{asset}} = 0.5 \log(2) + 0.5 \log(0.5) = 0$

## Simulation
- The provided Python code simulates the growth of both a rebalanced and an unbalanced portfolio
- The simulations demonstrate the advantage of continuous rebalancing in the long run, even in this simplified doubling-halving game

```python
#Simulation of Shannon's Demon
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the Black-Scholes call option pricing formula
def black_scholes_call(S, K, T, r, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    call_price = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    return call_price

# Define the Delta of a call option
def delta_call(S, K, T, r, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    delta = norm.cdf(d1)
    return delta

# Parameters
n_simulations = 100
n_periods = 1000  # Number of periods in each simulation
up_factor = 1.5  # Asset goes up by 50%
down_factor = 2/3  # Asset goes down by 33.33%%

# Function to simulate the process over time for visualization
def simulate_process_over_time(n_periods, up_factor, down_factor, n_simulations, balanced=True):
    all_paths = np.zeros((n_simulations, n_periods + 1))  # +1 to include the initial value
    for i in range(n_simulations):
        values = [1]  # Start each simulation with a value of 1
        stock = 0.5  # Initial stock proportion for balanced portfolio
        cash = 0.5  # Initial cash proportion for balanced portfolio
        for _ in range(n_periods):
            if np.random.rand() < 0.5:
                change_factor = up_factor
            else:
                change_factor = down_factor
            
            if balanced:
                # Update stock value and rebalance
                stock *= change_factor
                total = stock + cash
                stock = total / 2
                cash = total / 2
                values.append(stock + cash)
            else:
                # Update value without rebalancing
                values.append(values[-1] * change_factor)
                
        all_paths[i] = values

    return all_paths

# Generate paths
unbalanced_paths = simulate_process_over_time(n_periods, up_factor, down_factor, n_simulations, balanced=False)
balanced_paths = simulate_process_over_time(n_periods, up_factor, down_factor, n_simulations, balanced=True)

# Plotting
fig, axs = plt.subplots(2, 1, figsize=(10, 12))

# Unbalanced Portfolio Paths
for path in unbalanced_paths:
    axs[0].plot(path)
axs[0].set_title('Unbalanced Portfolio Value Over Time')
axs[0].set_xlabel('Time Period')
axs[0].set_ylabel('Portfolio Value')
axs[0].set_yscale('log')
# Balanced Portfolio Paths
for path in balanced_paths:
    axs[1].plot(path)
axs[1].set_title('Balanced Portfolio Value Over Time')
axs[1].set_xlabel('Time Period')
axs[1].set_ylabel('Portfolio Value')
axs[1].set_yscale('log')
plt.tight_layout()
plt.show()

```

## Dividends in the Black-Scholes Model
- The original Black-Scholes model assumes that the underlying asset pays no dividends
- However, many stocks pay dividends, which affects the price of the options
- Dividends are typically paid out of the company's earnings and distributed to shareholders
- When a stock pays a dividend, its price usually drops by the amount of the dividend on the ex-dividend date
- To account for dividends in the Black-Scholes model, we need to modify the formula

## Black-Scholes Formula with Dividends
The Black-Scholes formula for a European call option with dividends is given by:

$c(t) = S(t)e^{-q(T-t)} \cdot N(d_1) - Xe^{-r(T-t)} \cdot N(d_2)$

where:
- $q$: Continuous dividend yield (assumed to be constant)
- All other variables are the same as in the original Black-Scholes formula

The formula for a European put option with dividends is:

$p(t) = Xe^{-r(T-t)} \cdot N(-d_2) - S(t)e^{-q(T-t)} \cdot N(-d_1)$

The modified $d_1$ and $d_2$ are calculated as:

$d_1 = \frac{\ln(\frac{S(t)}{X}) + (r - q + \frac{\sigma^2}{2})(T-t)}{\sigma \sqrt{T-t}}$

$d_2 = d_1 - \sigma \sqrt{T-t}$

## Interpretation
- The dividend yield $q$ reduces the stock price growth rate, as the dividends are paid out of the company's earnings
- The term $e^{-q(T-t)}$ adjusts the current stock price $S(t)$ to account for the expected dividends paid during the option's lifetime
- As the dividend yield increases, the call option price decreases, and the put option price increases, all else being equal

```python
import numpy as np
from scipy.stats import norm

def black_scholes_dividend(call_put, S, X, T, r, q, sigma):
    """
    Calculates the price of a European call or put option with dividends using the Black-Scholes model.
    
    Parameters:
    call_put (str): 'call' or 'put'
    S (float): Current stock price
    X (float): Strike price
    T (float): Time to expiration (in years)
    r (float): Risk-free interest rate (continuous compounding)
    q (float): Dividend yield (continuous compounding)
    sigma (float): Volatility of the stock's returns
    
    Returns:
    float: Price of the option
    """
    
    d1 = (np.log(S / X) + (r - q + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    
    if call_put == 'call':
        price = S * np.exp(-q * T) * norm.cdf(d1) - X * np.exp(-r * T) * norm.cdf(d2)
    elif call_put == 'put':
        price = X * np.exp(-r * T) * norm.cdf(-d2) - S * np.exp(-q * T) * norm.cdf(-d1)
    else:
        raise ValueError("call_put must be either 'call' or 'put'")
    
    return price

# Example usage
call_price = black_scholes_dividend('call', 100, 100, 1, 0.05, 0.02, 0.2)
put_price = black_scholes_dividend('put', 100, 100, 1, 0.05, 0.02, 0.2)

print(f"Call price: {call_price:.2f}")
print(f"Put price: {put_price:.2f}")
```

## Score-Based Generative Modeling (Stephen's Research)

Score-based generative modeling is a recently developed approach to generating synthetic data, such as images, using stochastic differential equations (SDEs). The main idea behind this approach is to learn the score function, which is the gradient of the log-density function, and use it to guide the generation process.

### Forward SDE

In score-based generative modeling, the data generation process is modeled by a forward SDE:

$dX_t = f(X_t, t) dt + g(t) dW_t$

where:
- $X_t$ is the data point at time $t$
- $f(X_t, t)$ is the drift coefficient
- $g(t)$ is the diffusion coefficient
- $W_t$ is a standard Brownian motion

The drift and diffusion coefficients are chosen such that the marginal distribution of $X_T$ at the final time $T$ is a known prior distribution, such as a standard normal distribution.

### Reverse-Time SDE

To generate samples from the learned distribution, we need to solve the reverse-time SDE:

$dX_t = [f(X_t, t) - g(t)^2 \nabla_x \log p_t(X_t)] dt + g(t) d\bar{W}_t$

where:
- $\nabla_x \log p_t(X_t)$ is the score function, i.e., the gradient of the log-density of $X_t$ at time $t$
- $\bar{W}_t$ is a standard Brownian motion in the reverse-time direction

The reverse-time SDE starts from a sample from the prior distribution at time $T$ and gradually refines it to a sample from the learned data distribution at time $0$.
